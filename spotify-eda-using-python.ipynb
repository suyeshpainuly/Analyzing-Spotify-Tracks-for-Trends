```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('../input/spotify-datasets/tracks.csv')
df_tracks.head()
```

```python
# checking for null values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
#most popular songs
most_popular = df_tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending=False)
most_popular[:10]
```

```python
df_tracks.set_index('release_date', inplace=True) # inplace mutates the original df
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
#changing duration from ms to s
df_tracks['duration']=df_tracks['duration_ms'].apply(lambda x: round(x/1000)) # lambda is a small anonymous argument
df_tracks.drop('duration_ms', inplace=True, axis=1) # axis = 1 refers tp columns, 0 is rows
```

```python
df_tracks.duration.head()
```

```python
#finding correlation and plotting a heat map
corr_df= df_tracks.drop(['key', 'mode', 'explicit'], axis=1).corr(method='pearson')
plt.figure(figsize=(12,6))
heatmap=sns.heatmap(corr_df, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='mako', linewidths=1, linecolor='black')
heatmap.set_title('Correlation heatmap between different variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
plt.figure(figsize=(12,6))
sns.regplot(data = df_tracks, y = 'loudness', x = 'energy', color= 'r').set(title='Loudness vs Energy correlation')
```

```python
#to avoid overplotting, we can reduce the size of the data and plot it
sample_df=df_tracks.sample(int(0.005*len(df_tracks)))
plt.figure(figsize=(12,6))
sns.regplot(data = sample_df, y = 'loudness', x = 'energy', color= 'c').set(title='Loudness vs Energy correlation')
```

```python
plt.figure(figsize=(12,6))
sns.regplot(data = sample_df, y = 'popularity', x = 'acousticness', color= 'b').set(title='popularity vs acousticness correlation')
```

```python
# setting up to plot the number of songs in a year
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind='hist').set(title='Number of songs per year available on Spotify')
#the dip in 2021 is because the dataset is has data till 2021 only
```

```python
total_dr = df_tracks.duration
fig_dims = (20,8)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title = 'Duration of songs released per year')
plt.xlabel('Duration in seconds')
plt.ylabel('Year')
plt.xticks(rotation = 90)
```
